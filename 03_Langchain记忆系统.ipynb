### ConversationBufferWindowMemory

### ConversationTokenBufferMemory

```python
from langchain.memory import ConversationTokenBufferMemory
from langchain.chains import ConversationChain
#pip install transformers==4.53.2
```

```python
#智普ai
import os
from langchain_community.chat_models import ChatZhipuAI
key = 'b96d974eb5eb4648b9cdc76fbf423ddd.1bg7tnqu3cyIlpVg'
os.environ["ZHIPUAI_API_KEY"] = key
chat = ChatZhipuAI(
    model="glm-4",
    temperature=0.5, #模型温度  值在0-1之间 值越小 随机性越低
)
#创建记忆对象
memory = ConversationTokenBufferMemory(llm=chat,max_token_limit=50)
memory.save_context({"input":"Hi, how are you?"},{"output":"I am fine, thank you!"})
memory.save_context({"input":"What is your name?"},{"output":"My name is ChatZhipuAI!"})
memory.save_context({"input":"What is your age?"},{"output":"My age is 11111!"})
memory.save_context({"input":"What is your home?"},{"output":"My home is beijing!"})
```

```python
print(memory.buffer)
```

### ConversationSummaryBufferMemory

```python
from langchain.memory import ConversationSummaryBufferMemory
from langchain.chains import ConversationChain
```

```python
#创建记忆对象
memory = ConversationSummaryBufferMemory(llm=chat,max_token_limit=100)
memory.save_context({"input": "你好！我是小明"}, {"output": "您好，小明！有什么可以帮助您的？"})
memory.save_context({"input": "我喜欢编程，用 Python"}, {"output": "Python 很棒！您用它做什么？"})
memory.save_context({"input": "我在做一个聊天机器人"}, {"output": "酷！有什么特别功能吗？"})
memory.save_context({"input": "你记得我的职业吗？"}, {"output": "你是程序员，喜欢用Python，对吧？"})
```

```python
print(memory.buffer)
```

```python
#创建ConversationChain
conversation = ConversationChain(llm=chat,memory=memory,verbose=True) #verbose=True 打印调试过程
print(conversation.predict(input="你好,我的名字是小明"))
```
