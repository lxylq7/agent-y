```python
#智普ai
import os
from langchain_community.chat_models import ChatZhipuAI
from langchain.prompts import PromptTemplate
key = 'b96d974eb5eb4648b9cdc76fbf423ddd.1bg7tnqu3cyIlpVg'
os.environ["ZHIPUAI_API_KEY"] = key
chat = ChatZhipuAI(
    model="glm-4",
    temperature=0.5, #模型温度  值在0-1之间 值越小 随机性越低
)
from langchain_core.prompts import ChatPromptTemplate
#定义模板
sys_template = '你是数学老师,请以{style}风格回复'
user_template = '请解释一下{question}'
#创建对话模板
prompt_template = ChatPromptTemplate.from_messages([
    ('system',sys_template),
    ('human',user_template)
])
#填空变量
prompt = prompt_template.format_messages(style = '生动有趣',question = '勾股定理是什么？')
#获取响应
msg = chat.invoke(prompt)
print(msg.content)
```

### 输出格式化

```python
from langchain.output_parsers import StructuredOutputParser,ResponseSchema
from langchain.prompts import PromptTemplate
#智普ai
import os
from langchain_community.chat_models import ChatZhipuAI
from langchain.prompts import PromptTemplate
key = 'b96d974eb5eb4648b9cdc76fbf423ddd.1bg7tnqu3cyIlpVg'
os.environ["ZHIPUAI_API_KEY"] = key
chat = ChatZhipuAI(
    model="glm-4",
    temperature=0.5, #模型温度  值在0-1之间 值越小 随机性越低
)
#定义输出结构 自定义的格式
responsr_schemas = [
    ResponseSchema(name='name',description='人的姓名',type='string'),
    ResponseSchema(name='age',description='人的年龄',type='integer'),  
]
#创建输出解析器
output_parser = StructuredOutputParser.from_response_schemas(responsr_schemas)
#创建提示模板
template = """你是一个信息助手,请从以下文本中提取姓名和年龄,并以JSON格式返回
文本：{text}
{format_instructions}"""
prompt = PromptTemplate(template=template,
                        input_variables=["text"],
                        partial_variables={"format_instructions":output_parser.get_format_instructions()}) #解析器的格式说明
#填充输入
input_text = "张三今年25岁,来自北京"
filled_prompt = prompt.format(text=input_text)
#获取输出
resp = chat.invoke(filled_prompt)
parsed_output = output_parser.parse(resp.content)
print(parsed_output)
```

```python
from langchain.output_parsers import PydanticOutputParser
from pydantic import BaseModel,Field
from langchain.prompts import PromptTemplate
#定义输出模板
class Person(BaseModel):
    name: str = Field(description='人的姓名')
    age: int = Field(description='人的年龄')
#创建输出解析
output_parser = PydanticOutputParser(pydantic_object=Person)
#创建提示模板
template = """你是一个信息助手,请从以下文本中提取姓名和年龄,并以JSON格式返回
文本：{text}
{format_instructions}"""
prompt = PromptTemplate(
    template=template,
    input_variables=["text"],
    partial_variables={"format_instructions":output_parser.get_format_instructions()}
) #解析器的格式说明
#填充输入
input_text = "李四今年36岁,来自上海"
filled_prompt = prompt.format(text=input_text)
#获取输出
resp = chat.invoke(filled_prompt)
parsed_output = output_parser.parse(resp.content)
print(parsed_output)
```

### 链式调用

```python
from langchain.output_parsers import StructuredOutputParser,ResponseSchema
from langchain.prompts import PromptTemplate
#智普ai
import os
from langchain_community.chat_models import ChatZhipuAI
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
key = 'b96d974eb5eb4648b9cdc76fbf423ddd.1bg7tnqu3cyIlpVg'
os.environ["ZHIPUAI_API_KEY"] = key
chat = ChatZhipuAI(
    model="glm-4",
    temperature=0.5, #模型温度  值在0-1之间 值越小 随机性越低
)
#定义输出结构 自定义的格式
responsr_schemas = [
    ResponseSchema(name='answer',description='问题的答案',type='string'),
    ResponseSchema(name='confidence',description='答案的置信度',type='float')
]
#创建输出解析器
output_parser = StructuredOutputParser.from_response_schemas(responsr_schemas)
#创建提示模板
template = """你是一个数学老师,请用{style}风格回答以下问题,并以JSON格式返回答案和置信度:
问题：{question}
{format_instructions}"""
prompt = PromptTemplate(template=template,
                        input_variables=["text"],
                        partial_variables={"format_instructions":output_parser.get_format_instructions()}) #解析器的格式说明
#创建LLMchain
#llm_chain = LLMChain(llm=chat,prompt=prompt,output_parser=output_parser)
llm_chain = prompt | chat | output_parser #推荐写法
rs = llm_chain.invoke({
    'style' : '通俗易懂',
    'question' : '勾股定理是什么？'
})
print(rs)

    
```

### 流式输出

```python
from langchain.output_parsers import StructuredOutputParser,ResponseSchema
from langchain.prompts import PromptTemplate
#智普ai
import os
from langchain_community.chat_models import ChatZhipuAI
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
key = 'b96d974eb5eb4648b9cdc76fbf423ddd.1bg7tnqu3cyIlpVg'
os.environ["ZHIPUAI_API_KEY"] = key
chat = ChatZhipuAI(
    model="glm-4",
    temperature=0.5, #模型温度  值在0-1之间 值越小 随机性越低
)
#定义输出结构 自定义的格式
responsr_schemas = [
    ResponseSchema(name='answer',description='问题的答案',type='string'),
    ResponseSchema(name='confidence',description='答案的置信度',type='float')
]
#创建输出解析器
output_parser = StructuredOutputParser.from_response_schemas(responsr_schemas)
#创建提示模板
template = """你是一个数学老师,请用{style}风格回答以下问题,并以JSON格式返回答案和置信度:
问题：{question}
{format_instructions}"""
prompt = PromptTemplate(template=template,
                        input_variables=["text"],
                        partial_variables={"format_instructions":output_parser.get_format_instructions()}) #解析器的格式说明
#创建LLMchain
#llm_chain = LLMChain(llm=chat,prompt=prompt,output_parser=output_parser)
#llm_chain = prompt | chat | output_parser #推荐写法
llm_chain = prompt | chat
chunks = []
for chunk in llm_chain.stream({'style' : '通俗易懂','question' : '勾股定理是什么？'}): 
    chunks.append(chunk)
    print(chunk.content,end='',flush=True)

```
